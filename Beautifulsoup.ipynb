```python
import bs4 as bsoup

import urllib.request
```

```python
url = urllib.request.urlopen('https://www.amazon.in/Fire-Boltt-Smartwatch-Bluetooth-Calling-Assistance/dp/B09YV4MW2T/ref=sr_1_4?pd_rd_r=a99abf4a-fbe4-4a90-9d35-79f5907c06a6&pd_rd_w=4N0ry&pd_rd_wg=p42Ak&pf_rd_p=22a704c7-cdfe-40ab-ab4e-31cbf3ad0e01&pf_rd_r=G1RVDT2M385RHPZQXK4B&sr=8-4')
```

```python
bsoup = bsoup.BeautifulSoup(url,'lxml')
```

```python
bsoup
```

```python
tex_data = bsoup.get_text()

import re

pattern = re.compile(r'\n+|[^\w\s]+')

data = pattern.sub(' ',tex_data)
data

data =' '.join(data.split())

data
```
