```python
import bs4 as bs
import urllib.request
```

```python
src = urllib.request.urlopen('https://www.nytimes.com/').read()
```

```python
bsoup = bs.BeautifulSoup(src,'lxml')
```

```python
print(bsoup)
```

```python
# To print the Title of the websit

print(bsoup.title)
```

```python
# if we want to print the text inside the tag

print(bsoup.title.text)
```

```python
# if we want the all urls from the website

for link in bsoup.find_all('a'):
    print(link.get('href'))
```

```python
# if we want to print 1st paragraph in website

print(bsoup.p.text)
```

```python
# if we want to print all the paragraph

p_tags = bsoup.find_all('p')
for p in p_tags:
    print(p.get_text(),"\n")

```

<h3># Table Values Extraction From the Website</h3>

```python
import bs4 as bs
import urllib.request 

src = urllib.request.urlopen('https://www.nba.com/players').read()

bsoup = bs.BeautifulSoup(src,'lxml')

print(bsoup)
```

```python
tbl = bsoup.find('table') 
```

```python
import pandas as pd

tbl_rows = tbl.find_all(['tr'])

# Create an empty list to hold the rows
rows = []

for tr in tbl_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    rows.append(row)

# Convert the list of rows to a DataFrame
df = pd.DataFrame(rows)

# Print the resulting DataFrame
df.drop(index=0)
```
